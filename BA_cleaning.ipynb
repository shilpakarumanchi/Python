```python
import pandas as pd
import numpy as np
```

```python
#Importing British airways reviews dataset
df1 = pd.read_csv(r"C:\Users\shilp\OneDrive\Desktop\python_prjts\ba_reviews.csv")
df1
```

```python
#importing another data set that has information on Continent
df2 = pd.read_csv(r"C:\Users\shilp\OneDrive\Desktop\python_prjts\Countries.csv")
df2
```

```python
#joining the two data sets based on Country
df=pd.merge(df1,df2, left_on=['place'],right_on=['Country'])
df.info
```

```python
#Update display format
df.info()
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#Checking for null or empty values
df.isnull().sum()
```

```python
#Summary statics of numeric varaibles
df.describe()
# I notice that there are a number of "-1" ratings. Here "-1" was entered when there was no response provided.
# I would like to replace "-1" with the average rating within each service.
```

```python
# Replacing "-1" with the average rating within each service.
columns = ['seat_comfort','entertainment','cabin_staff_service','food_beverages','ground_service','value_for_money']
for col in columns:
    
    df[col] = df[col].astype('string').str.replace('-1', ' ')
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].astype(float).fillna(df.loc[df[col].notnull(), col].mean())
```

```python
#Summary statistics for each service 
df.describe()
#Now we donot have any negative raitngs
#The variable "rating" has a maximum value of 10 while all the services have a max. value of 5.
#I am not sure what rating represents here. Hence, I have calulated "average rating" based on the ratings of the 6 services ptovided.
```

```python
#I have calulated "average rating" based on the average of the 6 services ptovided.
df['avg_rating'] = (df['seat_comfort']+df['cabin_staff_service']+df['food_beverages']+df['ground_service']+df['value_for_money']+df['entertainment'])/6
```

```python
#Summary Statistics
df.describe()
#Inference:
#Among the services provided by the air line, highest rating is assoiated with cabin staff service, followed by ground service.
#Lowest rating is associated with food and beverages followed by entertainment.
```

```python
# Save the dataframe to a CSV file
df.to_csv('ba_reviews_clean.csv', index=False)
```

```python
#Checking for numbre of revies provided within each Continent
df2 = df.groupby(['Continent']).size().reset_index(name='Number of reviews')
df2
#Inference:
#Highest number of reviews were provided from Europe with 1009 reviews followed by North America with 168 reviews.
#Lowest number of reviews were provided from South America  with 5 reviews followed by Africa with 22 reviews. 
#There should be a minimum sample size of 30 inorder to be able to produce any inference.
```

```python
# Summary of ratings by Continent. Here I have not included Africa and South America, due to very small sample size.
df3 = df[df["Continent"].str.contains("Africa|South America") == False].groupby(['Continent'])[['seat_comfort',
       'cabin_staff_service', 'food_beverages', 'ground_service',
       'value_for_money', 'entertainment', 'avg_rating'
                ]].mean().sort_values(by="avg_rating",ascending=False)
df3
#Inference:
#Asia has the highest rating of 3.32, followed by Europe and NA with ratings of 2.92 and 2.88.
#Oceania (Australia & New Zealand) has the lowest average ratings of 2.58.
```

```python
#Summary ststistics by Seat type
df4 = df2.groupby(['seat_type'])[['seat_comfort',
       'cabin_staff_service', 'food_beverages', 'ground_service',
       'value_for_money', 'entertainment', 'avg_rating'
                ]].mean().sort_values(by="avg_rating",ascending=False)
df4
#Inference:
#On average the services provided in first class received higher ratings than the other class types.
#Economy class has the least ratings in all service categories.
```
